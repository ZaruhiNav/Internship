```python
import numpy as np
import pandas as pd

```

```python
data=pd.read_csv("data_sample.csv")
```

```python
data_numeric=data.select_dtypes(exclude="object")
```

```python
data_numeric.head()
```

```python
data_numeric=data_numeric.drop("RESPONSE_FLAG",axis=1)
```

```python
data_numeric.head()
```

```python
data_numeric_continuous=pd.DataFrame()
for i in data_numeric.columns:
    j=len(data_numeric[i].unique())
    if j>15:
        data_numeric_continuous=data_numeric_continuous.join(data_numeric[i],how='outer')
```

```python
data_numeric_continuous.head()
```

```python
data_numeric_discrete=pd.DataFrame()
for i in data_numeric.columns:
    j=len(data_numeric[i].unique())
    if j<=15:
        data_numeric_discrete=data_numeric_discrete.join(data_numeric[i],how='outer')
```
