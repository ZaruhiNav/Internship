```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_excel("addresses.xlsx")
```

```python
df1.head()
```

```python
df2=df1[df1["Community_Village_first"].isnull()]
```

```python
df1=df1.dropna()
df1.head()
```

```python
df1.Community_Village_first= [x.strip(' ') for x in df1.Community_Village_first]
```

```python
Community_list= df1.Community_Village_first.unique()
```

```python
print(Community_list)
```

```python
df2.head()
```

```python
for j in range(0,len(df2)):
    for i in Community_list:  
        if i in df2.loc[j,"address"]:
            df2.loc[j,"Community_Village_first"]=i
            break
        else:
            df2.loc[j,"Community_Village_first"]=0
```

```python
df2.head()
```

```python
df=df1.append(df2,sort=False)
```
